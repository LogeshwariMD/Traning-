```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
plt.hist(data, bins=range(12), density=True, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
mean_value = np.mean(data)
print("Simulated Mean:", round(mean_value, 2))
print("Theoretical Mean (n × p):", 10 * 0.5)
```

```python
if abs(mean_value - 5) < 0.5:
    print("Matches binomial expectation (n × p = 5).")
else:
    print("Does not match expectation; possible randomness in simulation.")
```
