Task 12: QQ Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
```

```python
np.random.seed(42)  
data = np.random.normal(loc=0, scale=1, size=100)  

print("Data generated successfully!")
```

```python
stats.probplot(data, dist="norm", plot=plt)
plt.title("QQ Plot - Normal Distribution")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.show()
```

```python
data_uniform = np.random.uniform(low=0, high=1, size=100)
data_exponential = np.random.exponential(scale=1, size=100)
plt.figure(figsize=(12, 5))

```

```python
plt.subplot(1, 2, 1)
stats.probplot(data_uniform, dist="norm", plot=plt)
plt.title("QQ Plot - Uniform Distribution")
```

```python
plt.subplot(1, 2, 2)
stats.probplot(data_exponential, dist="norm", plot=plt)
plt.title("QQ Plot - Exponential Distribution")
plt.tight_layout()
plt.show()
```
